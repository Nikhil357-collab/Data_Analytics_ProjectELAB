```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {"User-Agent": "Mozilla/5.0"}

jobs = []

roles = ["data analyst", "python developer"]
location = "India"

for role in roles:
    url = f"https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords={role}&location={location}&start=0"

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    job_cards = soup.find_all("div", class_="base-card")

    for job in job_cards:
        title = job.find("h3")
        company = job.find("h4")
        loc = job.find("span", class_="job-search-card__location")
        salary = job.find("span", class_="job-search-card__salary-info")

        jobs.append([
            title.text.strip() if title else None,
            company.text.strip() if company else None,
            loc.text.strip() if loc else None,
            salary.text.strip() if salary else "Not Disclosed",
            role
        ])

    time.sleep(2)

df = pd.DataFrame(jobs, columns=["Title", "Company", "Location", "Salary", "Role"])
df.head()
```

```python
skills_list = [
    "python", "sql", "excel", "machine learning",
    "power bi", "tableau", "java", "c++", "aws"
]

def extract_skills(text):
    text = str(text).lower()
    return [skill for skill in skills_list if skill in text]

df["Skills"] = df["Title"].apply(extract_skills)
df.head()
```

```python
df_exploded = df.explode("Skills")
df_exploded = df_exploded.dropna()
df_exploded = df.explode("Skills").dropna()
df_exploded.head()
```

```python
skill_city = df_exploded.groupby(["Location", "Skills"]).size().reset_index(name="Count")

top_skills = skill_city.sort_values(["Location", "Count"], ascending=[True, False])
top_skills.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pivot = skill_city.pivot(index="Skills", columns="Location", values="Count").fillna(0)

plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True)
plt.title("Skill Demand Heatmap")
plt.show()
```

```python
pivot = df_exploded.pivot_table(index="Skills", columns="Location", aggfunc="size", fill_value=0)

sns.heatmap(pivot, annot=True)
plt.show()
```

```python
pivot_role = df_exploded.pivot_table(index="Skills", columns="Role", aggfunc="size", fill_value=0)
pivot_role
```
